## Problems Encountered in the Map OSM Data

Main problems encountered are:
<ul>
<li>Street names are over-abbreviated and not consistent.</li>
<li>Postcode are incorrect or contain information that is not a postcode.</li>
</ul>

The sections below describe what these problems are in detail.

### Audit street names

Street names are stored as "v" value in "tag" with "k" value equal to "addr:street"

```python
%run audit_street_name.py $OSM_FILE
```

Auditing street names show the following common issues in the original osm file:
<ol>
  <li>Many street types are abbreviated, stored as all upper case or lower case. For example, "Road" is shown as "Rd.", "Rd" or "road". </li>
  <li>Directions are shown in various format: "N", "N.", "North", etc.</li>
  <li>Street names with numbers in it are in various format. For example, interstate highway names are displayed as "I 30", "Interstate Highway 30", etc.</li>
  <li>"Suite" in the street name is displayed inconsistently as "Suite", "suite", "#", etc. It should be converted to "Suite".</li>
</ol>

There are are multiple "special" issues with the street names:
<ol>
  <li>Street type is missing from the name. For example: "Arapaho" should be "Arapaho Road"</li>
  <li>"v" value is not a street name, but rather a house number or phone number</li>
  <li>"v" value contains not only street name but also other information</li>
</ol>

These special cases are first checked using the "id" attribute on openstreet map website. For example, if the node tag id is 1234567, its data can be viewed by typing "https://www.openstreetmap.org/node/1234567" in web browser.

These issues are fixed by running "clean_osm.py" to generate a new osm file and this file is used to convert to json foramt.

### Audit postcode

These issues are fixed by running "clean_osm.py" to create a new osm data file.

## Data Cleaning and Conversion

### Data Cleaning -- Generating a cleaned OSM data file

The street name and postcode have some special errors that need to be handled mannually. For example, the postcode field may not contain any valid post code. In these cases, the "id" value of the corresponding "way" or "node" tag is found first. Then the data is checked through the openstreet map website by typing "https://www.openstreetmap.org/<node|way>/<id>" in a web browser. The correct information is gathered and then fixed by running "clean_osm.py".<br>
Other issues such as over-abbreviated word in street names (type, direction, etc.) are also fixed through "clean_osm.py".<br>
Changes made are printed to the screen while running the script.

```python
# osm data after some manual fix
NEW_OSM_FILE = "../data/dallas_texas_new.osm"
%run clean_osm.py $OSM_FILE $NEW_OSM_FILE
```

### Converting OSM data file to JSON format

```python
# Converting data to json format
JSON_FILE = "../data/dallas_texas.json"
%run osm_to_json.py $NEW_OSM_FILE $JSON_FILE
```

Note:
<blockquote>
  The second level tag "tag" also has attribute "type". "type" is reserved for json document types, i.e. node, way. To avoid overwriting the document type, "type" attribute in the second-level tag is renamed to "tag_type".
</blockquote>
The created json file is then imported into database "osm" collection "dfw" in MongoDB using:
<blockquote>
mongoimport -d osm -c dfw --file ../data/dallas_texas.json
</blockquote>

```python
# number of "way" documents in dfw collection
db.dfw.find({"type" : "way"}).count()
```

```python
# number of unique contributors in dfw collection
len(db.dfw.distinct("created.user"))
```

```python
# Top 5 contributing user
list(db.dfw.aggregate([
    {"$group"   : {"_id"   : "$created.user", "count" : {"$sum" : 1}}},
    {"$project" : {"_id"   : "$_id", "count": "$count", "percentage" : {"$divide" : ["$count", 2675522]}}},
    {"$sort"    : {"count" : -1}},
    {"$limit"   : 5}]))
```

The top contributor contributed to 42.8% of the documents, which is much higher compared the 2nd contributor who contributed to only 3.7% of documents. Similar to the data observed in the sample project, the contributions of users seems extremely skewed

```python
# Number of users appearing only once (having 1 post)
list(db.dfw.aggregate([
    {"$group" : {"_id"   : "$created.user", "count" : {"$sum" : 1}}},
    {"$group" : {"_id"   : "$count", "num_of_users" : {"$sum" : 1}}},
    {"$sort"  : {"_id" : 1}},
    {"$limit" : 1}]))
```

# Other Ideas about the Dataset

```python
# Top 10 appearing amenities
list(db.dfw.aggregate([
    {"$match" : {"amenity" : {"$exists" : 1}}},
    {"$group" : {"_id"     : "$amenity", "count" : {"$sum" : 1}}},
    {"$sort"  : {"count"   : -1}},
    {"$limit" : 10}]))
```

```python
# Religions in Dallas area
list(db.dfw.aggregate([
    {"$match" : {"amenity" : "place_of_worship"}},
    {"$group" : {"_id"     : "$religion", "count"  : {"$sum" : 1}}},
    {"$sort"  : {"count"   : -1}}]))
```

```python
# Top 5 cuisines in Dallas area, including fast_food
list(db.dfw.aggregate([
    {"$match" : {"cuisine" : {"$exists" : True},
                 "amenity" : {"$regex"  : "restaurant|fast_food"}}},
    {"$group" : {"_id"     : "$cuisine", "count" : {"$sum" : 1}}},
    {"$sort"  : {"count"   : -1}},
    {"$limit" : 5}]))
```

```python
# Add a field "time" to created field and make it of type DATE
# Do not re-create "time" field if it already exists with the correct type Date
from datetime import datetime
for doc in db.dfw.find({"created.timestamp" : {"$exists" : True}}):
    if ( "time" not in doc["created"] or
            type(doc["created"]["time"]) is not datetime ):
        doc["created"]["time"] = datetime.strptime(doc["created"]["timestamp"],
                "%Y-%m-%dT%H:%M:%SZ")
        db.dfw.save(doc)
```

```python
# confirm the number of document with created.time as an Date object
db.dfw.find({"created.time" : {"$type" : 9}}).count()
```

```python
# Which year has the most data entries
list(db.dfw.aggregate([
    {"$group" : {"_id" : {"$year" : "$created.time"}, "count" : {"$sum" : 1}}},
    {"$project" : {"_id"   : "$_id", "count": "$count", "percentage" : {"$divide" : ["$count", 2675522]}}},
    {"$sort"  : {"count" : -1}}]))
```

Most of the data was entered in 2009; some changes have been done in 2015.
